```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# importing datasets
data_set = pd.read_csv('user_data.csv')
```

## x = data_set.iloc[:, [2, 3]].values # [2,3] array, x is 2d -- 2d array -- array of features
## y = data_set.iloc[:, 4].values........# 4 is single, y is 1d -- target

```python
# Extracting Independent and dependent Variable
x = data_set.iloc[:, [2, 3]].values 
y = data_set.iloc[:, 4].values

# Splitting the dataset into training and test set.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler  = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test  = scaler.transform(x_test)
```

```python
# x_test
```

## 2. Fitting a Decision-Tree algorithm to the Training set

```python
# Fitting Decision Tree classifier to the training set
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(x_train, y_train)
```

## 3. Predicting the test result

```python
# Predicting the test set result
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
# we have 100 features
len(y_pred)
```

## 4. Test accuracy of the result (Creation of Confusion matrix)

```python
# Creating the Confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

## 5. Visualizing the training set result:


```python
y_train, len(y_train)
```

```python
# Visualizing the training set result
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as mtp
import numpy as np

# Extract the training set and labels
x_set, y_set = x_train, y_train

# Create a meshgrid of points to plot the decision boundaries
x1, x2 = np.meshgrid(np.arange(start=x_set[:, 0].min() - 1, stop=x_set[:, 0].max() + 1, step=0.01),
                     np.arange(start=x_set[:, 1].min() - 1, stop=x_set[:, 1].max() + 1, step=0.01))


# x1, x2: These are the meshgrid arrays created using numpy.meshgrid.
# They represent the points in the feature space where the decision boundaries will be plotted.



# Specify enough colors for the number of unique classes
colors = ListedColormap(['purple', 'green', 'orange', 'blue'])  # Add more colors if needed

# Use the classifier to predict the classes for each point in the meshgrid

# and plot the decision boundaries
mtp.contourf(x1, x2, 
             classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha=0.75, 
             cmap=colors)

# Set plot limits
mtp.xlim(x1.min(), x1.max())
mtp.ylim(x2.min(), x2.max())

# mtp.xlim(-10,10)
# mtp.ylim(-10,10)

# Scatter plot of the actual training set points
for i, j in enumerate(np.unique(y_set)):
    mtp.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c=colors(i), label=str(j))

# Add title and labels to the plot
mtp.title('Decision Tree Algorithm (Training set)')
mtp.xlabel('Age')
mtp.ylabel('Estimated Salary')

# Display the legend
mtp.legend()

# Show the plot
mtp.show()
```

## 6. Visualizing the test set result

```python
# Visualizing the test set result
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as mtp
import numpy as np

x_set, y_set = x_test, y_test

x1, x2 = np.meshgrid(np.arange(start=x_set[:, 0].min() - 1, stop=x_set[:, 0].max() + 1, step=0.01),
                     np.arange(start=x_set[:, 1].min() - 1, stop=x_set[:, 1].max() + 1, step=0.01))

colors = ListedColormap(['purple', 'green', 'orange', 'blue'])  

mtp.contourf(x1, x2, 
             classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha=0.75, cmap=colors)

mtp.xlim(x1.min(), x1.max())
mtp.ylim(x2.min(), x2.max())

for i, j in enumerate(np.unique(y_set)):
    mtp.scatter(x_set[y_set == j, 0], 
                x_set[y_set == j, 1],
                c=colors(i), 
                label=str(j)
               )

mtp.title('Decision Tree Algorithm (Test set)')
mtp.xlabel('Age')
mtp.ylabel('Estimated Salary')
mtp.legend()
mtp.show()
```
